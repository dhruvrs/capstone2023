```python
rig['DATE'] = pd.to_datetime(rig['DATE'])
style.use('seaborn-talk')
plt.plot(rig['DATE'], rig['Texas Land'], linewidth=2, label="Texas")
plt.plot(rig['DATE'], rig["N. Mexico Land"], linewidth=2, label="New Mexico")
plt.plot(rig['DATE'], rig['Louisiana Land'], linewidth=2, label="Louisiana")
plt.plot(rig['DATE'], rig['Penn Land'], linewidth=2, label="Pennsylvania")
plt.plot(rig['DATE'], rig['N. Dakota Land'], linewidth=2, label="North Dakota")
plt.xlabel("Time")
plt.ylabel("Rig Count on Land")
plt.title("Rig Counts in Five States")
plt.legend()
plt.show()

# plt.plot(rig['DATE'], rig['Texas Land'], linewidth=2, color='#ed40b1', label="Texas")
# plt.plot(rig['DATE'], rig["N. Mexico Land"], linewidth=2, color="#1a911e", label="New Mexico")
# plt.plot(rig['DATE'], rig['Louisiana Land'], linewidth=2, color="#cc253b", label="Louisiana")
# plt.plot(rig['DATE'], rig['Penn Land'], linewidth=2, color="#2e35bf", label="Pennsylvania")
# plt.plot(rig['DATE'], rig['N. Dakota Land'], linewidth=2, color="#d18213", label="North Dakota")
```

```python
style.use('seaborn-talk')
crude['Date'] = pd.to_datetime(crude['Date'])
plt.plot(crude['Date'], crude['Crude Oil'], linewidth = 2)
plt.xlabel('Time')
plt.ylabel('Barrels of Crude(Thousand Barrels)')
plt.title('Total Crude Oil Stockpile')
```

```python
style.use('seaborn-talk')
gasStock['Date'] = pd.to_datetime(gasStock['Date'])
plt.figure(figsize=(10,5))
plt.plot(gasStock['Date'], gasStock['Gasoline Stock'], color = "#168233", linewidth = 1)
plt.xlabel('Time')
plt.ylabel('Barrels of Gasoline(Thousand Barrels)')
plt.legend()
plt.title('Total Gasoline Stock')
```

```python
henry['Date'] = pd.to_datetime(henry['Date'])
plt.figure(figsize=(10,5))
plt.plot(henry['Date'], henry['Henry Hub'], color = '#fc0356', linewidth = 1.2)
plt.xlabel('Time')
plt.ylabel('Price(Dollars per Million BTU)')
plt.title('Henry Hub Natural Gas Spot Price')
```

```python
natGas['Date'] = pd.to_datetime(natGas['year'])
style.use('seaborn-talk')
plt.figure(figsize=(10,5))
plt.plot(natGas['year'], natGas['imports'], linewidth = 3, label = 'Imports')
plt.plot(natGas['year'], natGas['exports'], linewidth = 3, label = 'Exports')
plt.xlabel('Time')
plt.ylabel('Trillion Cubic Feet')
plt.legend()
plt.title('US Net Exports and Imports of Gasoline')
```

```python
upRenew['Date'] = pd.to_datetime(upRenew['Date'])
finalRenew = upRenew.groupby(upRenew['Date'].dt.to_period('Y'), group_keys= True).sum()
finalRenew.insert(0, 'Date', range(1973, 2023))
for index, row in finalRenew.iterrows():
    if row['Date'] != 2000 and row['Date'] != 2010 and row['Date'] != 2020:
        finalRenew = finalRenew.drop(labels = index)
df = pd.read_csv('C:/Users/dhruv/Documents/School/Capstone/final renewables.csv')

dates = list(df.iloc[:,0])
woodEnergy = list(df.iloc[:,1])
bioMass = list(df.iloc[:,2])
renew = list(df.iloc[:,3])
x = np.arange(3)

width = 0.2
style.use('seaborn-talk')
plt.bar(x, woodEnergy, width, label='Wood Energy Production')
plt.bar(x+0.2, bioMass, width, label='Biomass Production')
plt.bar(x+0.4, renew, width,label='Total Renewable Production')
plt.xlabel('Years')
plt.ylabel('Trillion Btu')
plt.legend()
plt.xticks(x+width/2, dates)
plt.title("Renewable Energy Production")
```

```python
hydro = df.iloc[:,4]
geo = list(df.iloc[:,5])
wood = list(df.iloc[:,6])
waste = list(df.iloc[:,7])
totalBio = list(df.iloc[:,8])
totalRenew = list(df.iloc[:,9])

x = np.arange(3)

width = .075
style.use('seaborn-talk')
plt.figure(figsize=(10,5))
plt.bar(x, hydro, width, label='Hydroelectric Power')
plt.bar(x+width, geo, width, label='Geothermal Energy')
plt.bar(x+width*2, wood, width,label='Wood Energy')
plt.bar(x+width*3, waste, width,label='Waste Energy')
plt.bar(x+width*4, totalBio, width,label='Total Biomass Energy')
plt.bar(x+width*5, totalRenew, width,label='Total Renewable Energy')
plt.xlabel('Years')
plt.ylabel('Trillion Btu')
plt.legend()
plt.xticks(x+width/2, dates)
plt.title("Renewable Energy Consumption")

```
