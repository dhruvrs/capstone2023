```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
data = pd.read_csv('C:/Users/dhruv/Documents/School/Capstone/rig count by state.csv')
data
```

```python
data['DATE'] = pd.to_datetime(data['DATE'])
data
```

```python
plt.plot(data['DATE'], data['Texas Land'], linewidth=2, color='#ed40b1', label="Texas")
plt.plot(data['DATE'], data["N. Mexico Land"], linewidth=2, color="#1a911e", label="New Mexico")
plt.plot(data['DATE'], data['Louisiana Land'], linewidth=2, color="#cc253b", label="Louisiana")
plt.plot(data['DATE'], data['Penn Land'], linewidth=2, color="#2e35bf", label="Pennsylvania")
plt.plot(data['DATE'], data['N. Dakota Land'], linewidth=2, color="#d18213", label="North Dakota")
plt.xlabel("Time")
plt.ylabel("Rig Count on Land")
plt.title("Rig Counts in Five States")
plt.legend()
plt.show()
```
